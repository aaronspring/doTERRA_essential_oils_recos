```python
import pandas as pd

df = pd.read_csv("extracted_oils.csv")
df = df[df.url.str.contains("oil")]
df = df[df.name.notnull()]
df
```

```python
df.iloc[0]
```

```python
df["aromabeschreibung"] = df.aromabeschreibung.str.slice(1, -1).str.replace("'", "")
```

```python
df["hinweise_sichere_anwendung"] = (
    df.hinweise_sichere_anwendung.str.slice(1, -1).str.replace("',", "\n").str.replace("'", "")
)
```

```python
df["hauptnutzen"] = (
    df.hauptnutzen.str.slice(1, -1).str.replace("',", "\n").str.replace("'", "").to_list()
)
```

```python
df["anwendungsmoeglichkeiten"] = (
    df.anwendungsmoeglichkeiten.str.slice(1, -1)
    .str.replace("',", "\n")
    .str.replace("'", "")
    .to_list()
)
```

```python
df = df.fillna("")
```

```python
df["serialize"] = df.assign(
    serialize="Ätherisches Öl von dōTERRA **"
    + df.name
    + "** ("
    + df.lateinischer_name
    + ")\n"
    + "Produktbeschreibung: "
    + df.produktbeschreibung
    + "\n"
    + "Aromabeschreibung: "
    + df.aromabeschreibung
    + "\n"
    + "Hauptnutzen:\n "
    + df.hauptnutzen
    + "\n"
    + "Anwendungsmöglichkeiten:\n "
    + df.anwendungsmoeglichkeiten
    + "\n"
    + "Hinweise zur sicheren Anwendung:\n "
    + df.hinweise_sichere_anwendung
    + "\n"
).serialize.str.replace("()", "")
```
